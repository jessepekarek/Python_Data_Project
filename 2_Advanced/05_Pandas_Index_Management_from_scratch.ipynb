```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.index.name = 'job_index'
```

```python
df.head()
```

```python
df.index.dtype
```

```python

df.reset_index(drop=True, inplace=True)
df.head()
```

```python
job_median_salary =df.pivot_table(index='job_title_short', values='salary_year_avg', aggfunc='median')
job_median_salary.reset_index(inplace=True)
job_median_salary.set_index('salary_year_avg', inplace=True)
job_median_salary.head()
```

```python
pivot_table =df.pivot_table(index='company_name', values='salary_year_avg', aggfunc='median')
pivot_table.sort_index(inplace=True, ascending=False)
pivot_table.head()
```
